```python
import pandas as pd
```

```python
df = pd.read_csv('data/combined_shooting_census.csv')
df
```

```python
grouped = df.groupby(['state', 'city'])
grouped.mean()
```

```python
mean_age = grouped['age'].mean()
mean_age
```

```python
df['race'].unique()
```

```python
asian = df.loc[df['race'] == 'A',:]['race'].count()
white = df.loc[df['race'] == 'W',:]['race'].count()
hispanic = df.loc[df['race'] == 'H',:]['race'].count()
black = df.loc[df['race'] == 'B',:]['race'].count()
print(f'The amount of Asian people fatally shot is: {asian}')
print(f'The amount of White people fatally shot is: {white}')
print(f'The amount of Hispanic people fatally shot is: {hispanic}')
print(f'The amount of Black people fatally shot is: {black}')
```

```python
race_grouping = df.loc[:, ["id", "race"]]
race_grouping = race_grouping.groupby(["race"])["id"].count()
race_grouping = pd.DataFrame(race_grouping)
race_grouping.head(10)
```

```python
unarmed = df.loc[(df['armed'] == 'unarmed') | (df['armed'] == 'NaN'), 'armed'].count()
count_armed = df['armed'].count() - unarmed
print(f'The number of people armed when they were fatally shot is: {count_armed}')
```
