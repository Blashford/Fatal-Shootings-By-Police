```python
import pandas as pd
```

```python
shooting_df = pd.read_csv('data/fatal-police-shootings-data.csv')
shooting_df
```

```python
cali_shot = shooting_df.loc[shooting_df['state'] == 'CA', :]
cali_shot
```

```python
census_df = pd.read_csv('data/full_census_data.csv')
census_df = census_df.drop(columns=['Unnamed: 0', 'Unnamed: 0.1'])
census_df
```

```python
cali_census = census_df.loc[census_df['Geographic Area'] == 'CA', :]
cali_census
```

```python
cali_census = cali_census.rename(columns = {'City': 'city'})
cali_census
```

```python
cali_merged = cali_shot.merge(cali_census, how = 'left', on='city')
cali_merged
```

```python
cali_merged = cali_merged.drop_duplicates(['id'])
cali_merged
```

```python
cali_merged.to_csv('data/cali_combined.csv')
```
