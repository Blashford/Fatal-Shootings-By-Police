```python
import pandas as pd
```

```python
df = pd.read_csv('data/cali_combined.csv')
df = df.drop(columns='Unnamed: 0')
df
```

```python
grouped = df.groupby('city')
grouped['age'].mean()
```

```python
race_grouping = df.loc[:, ["id", "race"]]
race_grouping = race_grouping.groupby(["race"])["id"].count()
race_grouping = pd.DataFrame(race_grouping)
race_grouping.head(10)
```

```python
unarmed = df.loc[(df['armed'] == 'unarmed') | (df['armed'] == 'NaN'), 'armed'].count()
count_armed = df['armed'].count() - unarmed
print(f'The number of people armed when they were fatally shot is: {count_armed}')
```
