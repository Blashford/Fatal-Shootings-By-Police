```python
import pandas as pd
```

```python
shooting_df = pd.read_csv('data/fatal-police-shootings-data.csv')
shooting_df
```

```python
census_df = pd.read_csv('data/full_census_data.csv')
census_df = census_df.drop(columns=['Unnamed: 0', 'Unnamed: 0.1'])
census_df
```

```python
census_df = census_df.rename(columns={'City': 'city', 'Geographic Area': 'state'})
census_df
```

```python
combined_df = shooting_df.merge(census_df, how='left', on=['state', 'city'])
combined_df
```

```python
combined_df = combined_df.drop_duplicates(['id'])
combined_df
```

```python
combined_df.to_csv('data/combined_shooting_census.csv', index = False)
```
